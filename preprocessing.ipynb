```python
import nltk

nltk.download('gutenberg')


nltk.corpus.gutenberg.fileids()
```

```python
Book1 = nltk.corpus.gutenberg.words('austen-emma.txt')
First500 = Book1[0:500]
```

```python
# step 1 - Remove all the puntuation and non words

norm_1 = [word for word in First500 if word.isalpha()]
norm_1
```

```python
# step 2 - Remove stopwords

from nltk.corpus import stopwords
nltk.download('stopwords')

norm_2 = [word for word in norm_1 if word not in stopwords.words('english')]

norm_2
```

```python
# step 3 - convert to lowercase

norm_3 = [word.lower() for word in norm_2]
norm_3
```

```python
# step 4 - use stemmer

stemmmer = nltk.stem.PorterStemmer()
norm_4 = [stemmmer.stem(word) for word in norm_3]

for i in range(len(norm_3)):
  if norm_3[i] != norm_4[i]:
    print(norm_3[i], " : ", norm_4[i])
```

```python
# step 5 - use lemmatization

lemmmatizer = nltk.stem.WordNetLemmatizer()
nltk.download('wordnet')

norm_5 = [lemmmatizer.lemmatize(word) for word in norm_4]

for i in range(len(norm_4)):
  if norm_4[i] != norm_5[i]:
    print(norm_4[i], " : ", norm_5[i])
```
